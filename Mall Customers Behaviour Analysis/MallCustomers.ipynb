```python
# Assuming df is your DataFrame
df_Gender = df.groupby(['Gender', 'Age'])[['Annual Income (k$)']].sum()
df_Gender = df_Gender.sort_values('Annual Income (k$)')

# Resetting the index to make plotting easier
df_Gender = df_Gender.reset_index()

# Plotting the histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plotting histogram for Male
axes[0].hist(df_Gender[df_Gender['Gender'] == 'Male']['Age'], bins=50, weights=df_Gender[df_Gender['Gender'] == 'Male']['Annual Income (k$)'], alpha=0.7, label='Male')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Annual Income (k$)')
axes[0].set_title('Male')
axes[0].legend()

# Plotting histogram for Female
axes[1].hist(df_Gender[df_Gender['Gender'] == 'Female']['Age'], bins=50, weights=df_Gender[df_Gender['Gender'] == 'Female']['Annual Income (k$)'], alpha=0.7, label='Female')
axes[1].set_xlabel('Age')
axes[1].set_title('Female')
axes[1].legend()

fig.suptitle('Histogram of Annual Income by Age and Gender')
plt.show()
```

```python
# Assuming df is your DataFrame
df_Gender = df.groupby(['Gender', 'Age'])[['Spending Score (1-100)']].sum()
df_Gender = df_Gender.sort_values('Spending Score (1-100)')

# Resetting the index to make plotting easier
df_Gender = df_Gender.reset_index()

# Plotting the histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plotting histogram for Male
axes[0].hist(df_Gender[df_Gender['Gender'] == 'Male']['Age'], bins=50, weights=df_Gender[df_Gender['Gender'] == 'Male']['Spending Score (1-100)'], alpha=0.7, label='Male')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Spending Score (1-100)')
axes[0].set_title('Male')
axes[0].legend()

# Plotting histogram for Female
axes[1].hist(df_Gender[df_Gender['Gender'] == 'Female']['Age'], bins=50, weights=df_Gender[df_Gender['Gender'] == 'Female']['Spending Score (1-100)'], alpha=0.7, label='Female')
axes[1].set_xlabel('Age')
axes[1].set_title('Female')
axes[1].legend()

fig.suptitle('Histogram of Spending Score (1-100) by Age and Gender')
plt.show()
```
